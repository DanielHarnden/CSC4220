## <font color="green">Data preprocessing and cleaning</font>

#### <font color="green">Now we will process the texts to remove punctuations, convert to lowercase, remove [stopwords](https://www.meaningcloud.com/developer/resources/doc/models/1.0/models/stopwords), and [stem the words](https://www.geeksforgeeks.org/python-stemming-words-with-nltk/)</font>

```python
# Download the required libraries
import re   # regular expression library
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
# Create a list to contain cleaned and processed texts from all instances
corpus = []

for i in range(0, len(dataset)): # iterates through all texts
    # substitute anything that's not a letter (i.e.punctuations) with space
    text = re.sub('[^a-zA-Z]', ' ', dataset['text'][i])
    
    # convert to lowercase
    text = text.lower()
    
    # split the text into a list of individual words
    text = text.split()
    ps = PorterStemmer() # python library for stemming words
    
    # create list to store the stemmed words
    stemmedWord = []
    for word in text:
        if not word in set(stopwords.words('english')): # ignore stopwords
            stemmedWord.append(ps.stem(word))
    
    # join the words in stemmedWord list back together to form a string (with spaces between them)
    text = ' '.join(stemmedWord)
    
    # add the processed text to corpus 
    corpus.append(text)
```

```python
print(corpus)
```

## <font color="green">Feature extraction</font>

### <font color="green">Creating the [Bag of Words](https://machinelearningmastery.com/gentle-introduction-bag-words-model/) model</font>

```python
from sklearn.feature_extraction.text import CountVectorizer
# instantiate a model, and fit the processed text (corpus)
BoW_model = CountVectorizer()
X = BoW_model.fit_transform(corpus).toarray()

# Set y as the spam column
y = dataset['label'].values
```

### <font color="green">Splitting the dataset into the Training set and Test set</font>

**<font color="green" size=3>Write the code to split the dataframe into train and test sets (chapter 2 of textbook train_test_split on page 53)</font>**

**<font color="red" size=3>  Daniel Harnden's Code:</font>**

```python
# Import library for train_test_split
from sklearn.model_selection import train_test_split
# Split training and test data into 80/20 split (test_size=0.2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### <font color="green">Training the Naive Bayes model on the Training set</font>

```python
from sklearn.naive_bayes import GaussianNB
# create our Gaussian Naive Bayes model and fit it to our training data
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

### <font color="green">Predicting the Test set results</font>

```python
# predict the texts in X_test as spam or ham
y_pred = classifier.predict(X_test)
# Display a side by side comparison of the predicted label and true label
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### <font color="green">Evaluating the model</font>

```python
# Compute model accuracy
from sklearn.metrics import accuracy_score
BoW_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model with Bag of Words feature extraction: ", BoW_accuracy)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# producing a normalized Confusion Matrix
cm = confusion_matrix(y_test, y_pred, normalize='true')
print(cm)
```

### <font color="green">Let's produce a nice looking confusion matrix</font>

```python
import itertools

# plot confusion matrix function
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
class_names = ['not spam','spam']
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()

# We can get each value of the confusion matrix individually
# Let's assume the spam class is the positive class
print("True positives:",cnf_matrix[1,1])
print("True negatives:",cnf_matrix[0,0])
print("False positives:",cnf_matrix[0,1])
print("False negatives:",cnf_matrix[1,0])
```

**<font color="green" size=3>Now write the code to implement TF-IDF feature extraction method. Train with the GaussianNB model and calculate the accuracy and produce a confusion matrix.</font>**

**<font color="red" size=3>  Daniel Harnden's Code:</font>**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Uses TF-IDF feature extraction
vect = TfidfVectorizer()
Xtf_idf = vect.fit_transform(corpus).toarray()

# Creates a training and testing set using the TF-IDF extracted values
tfX_train, tfX_test, tfy_train, tfy_test = train_test_split(Xtf_idf, y, test_size=0.2)

# Trains with the GaussianNB model
tf_idfClassifier = GaussianNB()
tf_idfClassifier.fit(tfX_train, tfy_train)

# Calculates the accuracy
tfy_pred = tf_idfClassifier.predict(X_test)
tf_idfAcc = accuracy_score(tfy_test, tfy_pred)
print("Accuracy of the model with TF-IDF feature extraction: ", tf_idfAcc)

# Produces a confusion matrix
tfcm = confusion_matrix(tfy_test, tfy_pred, normalize='true')
print(tfcm)
```

**<font color="green" size=3>With features extracted from the TF-IDF method, implement a logistic regression model and calculate accuracy and create confusion matrix.</font>**

**<font color="red" size=3>  Daniel Harnden's Code:</font>**

```python
from sklearn.linear_model import LogisticRegression

# Trains using the Logicstic Regression model
logReg = LogisticRegression(random_state=0).fit(tfX_train, tfy_train)

# Calculates the accuracy
lry_pred = logReg.predict(tfX_test)
logRegAcc = accuracy_score(tfy_test, lry_pred)
print("Accuracy of the Logistic Regression model: ", logRegAcc)

# Produces a confusion matrix
lrcm = confusion_matrix(tfy_test, lry_pred, normalize='true')
print(lrcm)
```

**<font color="green" size=3>In the box below, write the code to generate the ROC curves for both the Naive Bayes and the  logistic regression classifier on the same graph.</font>**

**<font color="red" size=3>  Daniel Harnden's Code:</font>**

```python
from sklearn.metrics import roc_curve

# Calculates the y_true
tfprobs = tf_idfClassifier.predict_proba(tfX_test)
lrprobs = logReg.predict_proba(tfX_test)

# Calculates the y_score
tfpreds = tfprobs[:,1]
lrpreds = lrprobs[:,1]

# Generates the ROC curves
tffpr, tftpr, threshold = roc_curve(tfy_test, tfpreds)
lrfpr, lrtpr, threshold = roc_curve(tfy_test, lrpreds)

# PLots the ROC curves and labels the graph
plt.title('Receiver Operating Characteristic')
plt.plot(lrfpr, lrtpr, threshold, label="Logistic Regression")
plt.plot(tffpr, tftpr, threshold, label="Gaussian NB")
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend()
plt.show()
```

**<font color="green" size=3>In the box below, write the code to compute and print the AUC for both curves.</font>**

**<font color="red" size=3>  Daniel Harnden's Code:</font>**

```python
from sklearn.metrics import auc

# Computes AUC for Gaussian NB and LR
print("Gaussian NB AUC: " + str(auc(tffpr, tftpr)))
print("Logistic Regression AUC: " + str(auc(lrfpr, lrtpr)))
```
